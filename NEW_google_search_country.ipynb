```python
country_name=input("ENTER THE COUNTRY NAME:")
print(country_name)
```

```python
import os
from crewai import Agent, Task, Crew
# Importing crewAI tools
from crewai_tools import (
    SerperDevTool,
    WebsiteSearchTool,
    DirectoryReadTool,
    FileReadTool


)

import json
import os

import requests
from langchain.tools import tool

import os
from crewai import Agent, Task, Crew
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import load_tools

search_tool = SerperDevTool()
web_rag_tool = WebsiteSearchTool()
file_tool = FileReadTool()



# Create agents
researcher = Agent(
    role='Rules regulator',
    goal=f'Extract basic rules and culture from {country_name}',
    backstory="""You're a Rules and culture regulator at a large company.
    You're responsible for extracting datas from google and show all the basic rules and culture of a country.
    You're currently working on a project to analyze the
    rules and rugulations of a country.""",
    tools=[search_tool,web_rag_tool],  # Optional, defaults to an empty list
    #llm=my_llm,  # Optional
    #function_calling_llm=my_llm,  # Optional
    #max_iter=15,  # Optional
    #max_rpm=None, # Optional
    #verbose=True,  # Optional
    #allow_delegation=True,  # Optional
    #step_callback=my_intermediate_step_callback,  # Optional
    #cache=True  # Optional
)

writer = Agent(
    role='rules Writer',
    goal=f'Craft engaging recent rules and cultural decentces of {country_name}',
    backstory='A skilled writer with a passion for cultural habited.',
    tools=[ file_tool, web_rag_tool],
    verbose=True
)



# Define tasks
research = Task(
    description='Research the latest rules and regulations implemented.',
    expected_output='A summary of the top rules and cultural habites of that country.',
    agent=researcher,
)

write = Task(
    description='Write the rules and regualations of that country and a brief about their culture.',
    expected_output='A point-wise output with all nessary rules and regulations of the country.',
    agent=writer

)

# Assemble a crew
crew = Crew(
    agents=[researcher, writer],
    tasks=[research, write],
    verbose=2
)

# Execute tasks
results=crew.kickoff()
```
